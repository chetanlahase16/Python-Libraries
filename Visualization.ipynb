```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

# Line Chart

```python
X = np.array([0,7])
Y = np.array([0,89])

plt.plot(X,Y)
plt.show()
```

```python
point = np.array([20,60,40,55,75,80])
plt.plot(point, linestyle = "dotted")        #we can also use dashed
plt.show()
```

# Plot Chart

```python
x = np.array(["Chetan", "Pratik", "Mahesh"])
y = np.array([34, 89, 20])
plt.bar(x,y)
plt.show()      # In bar chart we work with one string and one numerical field
```

# Scatter Plot

```python
x1 = np.array([20,40,60,80,100])
y1 = np.array([10,6.6,5.6,7.2,9.8])
```

```python
plt.scatter(x1,y1)
plt.show()
```

# Pie Chart

```python
p = np.array([32, 10,48,65])
plt.pie(p)
```

# Histogram

```python
h1 = np.random.normal(120,190,560)
plt.hist(h1)
```
